```python
import matplotlib.pyplot as plt
import PIL
import urllib
```

```python
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/2/29/Himalayas%2C_Ama_Dablam%2C_Nepal.jpg/640px-Himalayas%2C_Ama_Dablam%2C_Nepal.jpg"
```

```python
image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open (image_data)
np_image = np.array(image)
```

```python
plt.imshow(np_image)
```

```python
np_image.shape
```

```python
plt.imshow(np_image[200:500,100:500,0:4])
```

```python
image_red_only = np_image.copy()
image_red_only[:,:,1] = 0
image_red_only[:,:,2] = 0
```

```python
image_red_only2 = np_image.copy()
image_red_only2[200:300,:,1] = 0
image_red_only2[200:300,:,2] = 0
```
